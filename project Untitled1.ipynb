```python
df = pd.read_csv("Amazon - Movies and TV Ratings.csv")
df1 = df.drop(["user_id"], axis=1)
column = df1.columns
avg = pd.DataFrame()
for i in column:
    
    s={"movie_name":[i],"sum_rating":[df[i].dropna().sum()],"avg_rating":[df[i].dropna().mean()],"num_ratings":[df[i].count()]}
    s1=pd.DataFrame(s)
    avg = avg.append(s1, ignore_index=True)
```

### Movies with maximum average ratings

```python
avg.sort_values(by = "avg_rating", ascending = False).head(10)
```

### Movies with maximum views/ratings

```python
avg.sort_values(by = "num_ratings", ascending = False)
```

### Average Rating of each movie

```python
avg[["movie_name","avg_rating"]]
```

### Top movies with with maximum ratings

```python
avg.sort_values(by = ["num_ratings","avg_rating"], ascending = [False,False]).head()
```

### Top movies with least audience

```python
avg.sort_values(by = ["num_ratings","avg_rating"], ascending = [True,False]).head()
```

```python
import surprise
```

```python
from surprise import Reader
from surprise import Dataset
from surprise import SVD
from surprise.model_selection import train_test_split
```

```python
reader = Reader(rating_scale=(-1,10))


data = Dataset.load_from_df(melt_df.fillna(0), reader=reader)
```

###Divide the data into training and test data

```python
trainset, testset = train_test_split(data, test_size=0.25)
```

```python
algo = SVD()
```

###Building a recommendation model on training data

```python
algo.fit(trainset)
```

###Make predictions on the test data

```python
predict= algo.test(testset)
```

```python
from surprise.model_selection import cross_validate
```

```python
cross_validate(algo,data,measures=['RMSE','MAE'],cv=3,verbose=True)
```
